```python
from sklearn.ensemble import RandomForestClassifier

# Define features (X) and target (y)
X = df.drop(['diagnosis', 'Unnamed: 32', 'id'], axis=1)  # Exclude non-feature columns
y = df['diagnosis']

# Initialize and train a RandomForestClassifier
rf_classifier = RandomForestClassifier(random_state=100) # Set random_state for reproducibility
rf_classifier.fit(X, y)

# Get feature importances
feature_importances = rf_classifier.feature_importances_

# Create a DataFrame to store feature names and their importances
feature_importance_df = pd.DataFrame({'Feature': X.columns, 'Importance': feature_importances})

# Sort by importance in descending order
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# Print or visualize the feature importances
print(feature_importance_df)



plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=feature_importance_df)
plt.title('Feature Importance (Random Forest)')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,f1_score,fbeta_score,confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.neighbors import RadiusNeighborsClassifier
```

```python
##independent and dependent features
X = df.drop(['diagnosis', 'Unnamed: 32', 'id'], axis=1)
Y = df['diagnosis']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.30, random_state=42)
```

```python
treemodel=RandomForestClassifier()
treemodel.fit(X_train,Y_train)
```

```python
f1score=f1_score(Y_test,treemodel.predict(X_test),average='weighted')
f1score
classificationsreport=classification_report(Y_test,treemodel.predict(X_test))
print(classificationsreport)
fbeta=fbeta_score(Y_test,treemodel.predict(X_test),average='weighted',beta=0.5)
fbeta
```

```python
svm=svm.SVC()
svm.fit(X_train,Y_train)
```

```python
f1score=f1_score(Y_test,svm.predict(X_test),average='weighted')
f1score
classificationsreport=classification_report(Y_test,svm.predict(X_test))
print(classificationsreport)
fbeta=fbeta_score(Y_test,svm.predict(X_test),average='weighted',beta=0.5)
fbeta
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
radiusclassifier=RadiusNeighborsClassifier(radius=10)
radiusclassifier.fit(X_train,Y_train)
```

```python
confusion_matrix(Y_test, radiusclassifier.predict(X_test))
f1score=f1_score(Y_test,radiusclassifier.predict(X_test),average='weighted')
f1score
classificationsreport=classification_report(Y_test,radiusclassifier.predict(X_test))
print(classificationsreport)
fbeta=fbeta_score(Y_test,radiusclassifier.predict(X_test),average='weighted',beta=0.5)
fbeta
```

```python
model_accuracies = {
    'Random Forest': 0.97,
    'SVM': 0.93,
    'Radius Neighbors': 0.71
}

# Create a bar plot
plt.figure(figsize=(8, 6))
models = list(model_accuracies.keys())
accuracies = list(model_accuracies.values())
plt.bar(models, accuracies, color=['skyblue', 'salmon', 'lightgreen'])
plt.xlabel("Models")
plt.ylabel("Accuracy Score")
plt.title("Model Comparison")
plt.ylim(0, 1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
for i, v in enumerate(accuracies):
    plt.text(i, v + 0.01, f"{v:.2f}", ha='center')

plt.show()
```
